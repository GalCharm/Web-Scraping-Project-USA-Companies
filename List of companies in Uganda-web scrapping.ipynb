```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_Uganda'
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
soup.find('table')
```

```python
table = soup.find('table')
```

```python
print(table)
```

```python
table.find_all('th')
```

```python
tables_titles = table.find_all('th')
tables_titles
```

```python
list_table_titles = [title.text.strip() for title in tables_titles]
print(list_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = list_table_titles)
df
```

```python
table.find_all('tr')
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
